```python
import numpy as np
import pandas as pd
```

```python
file = pd.read_csv("YouTubeDataset_withChannelElapsed.csv")
```

```python
file
```

```python
file = file[['videoId', 'videoLikeCount', 'videoDislikeCount', 'VideoCommentCount', 'totalviews/channelelapsedtime', 'views/elapsedtime', 'elapsedtime', 'channelViewCount']]
```

```python
file = file.groupby('videoId').agg({
    'videoLikeCount': 'sum',
    'videoDislikeCount': 'sum',
    'VideoCommentCount': 'sum',
    'totalviews/channelelapsedtime': 'sum',
    'views/elapsedtime': 'sum',
    'elapsedtime': 'sum',
    'channelViewCount': 'sum'
}).reset_index()
```

```python
likes = np.round(file['videoLikeCount'].mean(), 4)
print(likes)
```

```python
df_clean = file[file['totalviews/channelelapsedtime'] != 0]

engament = np.round(((df_clean['videoLikeCount'] + df_clean['videoDislikeCount'] + df_clean['VideoCommentCount']) / df_clean['totalviews/channelelapsedtime']).mean(), 4)
print(engament)
```

```python
storages = np.round(((200 * 1024 + 500) / (1024 * 1024)), 4)
print(storages)
```

```python
data_trans =  np.round(storages * (file['videoLikeCount'].sum() + file['videoDislikeCount'].sum() + file['VideoCommentCount'].sum()), 4)
print(data_trans)
```

```python
total_views = np.round(((file['views/elapsedtime'] * file['elapsedtime']).sum() / (5*365*24*50)), 4)
print(total_views)
```

```python
data = np.round(total_views *storages, 4)
print(data)
```
